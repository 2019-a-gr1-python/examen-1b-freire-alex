```python
def get_different_colors(n):
    max_value = 16581375 
    interval = int(max_value / n)
    colors = [hex(I)[2:].zfill(6) for I in range(0, max_value, interval)]
    
    return [(int(i[:2], 16), int(i[2:4], 16), int(i[4:], 16)) for i in colors]
```

```python
len(mapache)
```

```python
get_different_colors(len(mapache))
```
